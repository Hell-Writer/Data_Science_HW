# Задача 2

```python
cnt=0
n_sim = 10000
np.random.seed(swag)
random.seed(swag)
atos_stats, partos_stats, aramis_stats = [], [], []
rand_size = n_sim * 2000 #2000, чтобы хватило на список из тысячи независимых значений для пар Партос-Арамис и Атос-Арамис
a_rand_list = np.random.binomial(n=1, p=0.49, size=rand_size) #Генерирую значения сразу списком, потому что так в 4 раза быстрее
b1_rand_list = np.random.binomial(n=1, p=0.09, size=rand_size)
b2_rand_list = np.random.binomial(n=1, p=0.74, size=rand_size)

def casino_a(wealth, cnt):
    gain = a_rand_list[cnt]
    gain = 2 * int(gain) - 1 #Спонтанное желание не использовать оператор if
    return wealth + gain

def casino_b(wealth, cnt):
    if wealth % 3 == 0:
        gain = b1_rand_list[cnt]
    else:
        gain = b2_rand_list[cnt]
    gain = 2 * int(gain) - 1
    return wealth + gain

def casino_aramis(wealth, cnt):
    casino_a_bool = random.getrandbits(1)
    if casino_a_bool == 1:
        return casino_a(wealth, -cnt) #Тут стоит минус, чтобы случайные величины не стали зависимыми 
    else:
        return casino_b(wealth, -cnt)
    
for i in range(n_sim):
    atos_wealth, partos_wealth, aramis_wealth = 1000, 1000, 1000
    atos_wealth_dinamics = [1000]
    partos_wealth_dinamics = [1000]
    aramis_wealth_dinamics = [1000]
    for j in range(1000):
        atos_wealth = casino_a(atos_wealth, cnt)
        partos_wealth = casino_b(partos_wealth, cnt)
        aramis_wealth = casino_aramis(aramis_wealth, cnt)
        atos_wealth_dinamics.append(atos_wealth)
        partos_wealth_dinamics.append(partos_wealth)
        aramis_wealth_dinamics.append(aramis_wealth)
        cnt+=1
    atos_stats.append(atos_wealth_dinamics)
    partos_stats.append(partos_wealth_dinamics)
    aramis_stats.append(aramis_wealth_dinamics)

avg_atos_dinamics = [sum(item)/len(item) for item in list(list(map(list, np.array(atos_stats, dtype=int).T)))]    
avg_partos_dinamics = [sum(item)/len(item) for item in list(list(map(list, np.array(partos_stats, dtype=int).T)))]   
avg_aramis_dinamics = [sum(item)/len(item) for item in list(list(map(list, np.array(aramis_stats, dtype=int).T)))]

visible_ids = np.random.randint(0, n_sim, 10, dtype=int)
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 15))

line12 = ax1.plot(avg_atos_dinamics, linewidth=3.0, color='r', label='b')
ax1.set_ylabel('Богатство Атоса', fontsize=18)
ax1.set_title('Изменения богатства Атоса')

ax2.plot(avg_partos_dinamics, linewidth=3.0, color='r')
ax2.set_ylabel('Богатство Портоса', fontsize=18)
ax2.set_title('Изменения богатства Портоса')

ax3.plot(avg_aramis_dinamics, linewidth=3.0, color='r')
ax3.set_ylabel('Богатство Арамиса', fontsize=18)
# Чтобы не писать три раза одно и то же, здесь и далее все оси абсцисс отмечены один раз - в самом низу
ax3.set_xlabel('Номер похода в казино', fontsize=18)
ax3.set_title('Изменения богатства Арамиса')

for item in visible_ids:
    ax1.plot(atos_stats[item], linewidth=0.2, color='blue', label='a')
    ax2.plot(partos_stats[item], linewidth=0.2, color='blue')
    ax3.plot(aramis_stats[item], linewidth=0.2, color='blue')
ax1.legend(['Усреднённое богатство Атоса', 'Динамика богатства Атоса в случайной симмуляции'])
ax2.legend(['Усреднённое богатство Портоса', 'Динамика богатства Портоса в случайной симмуляции'])
ax3.legend(['Усреднённое богатство Арамиса', 'Динамика богатства Арамиса в случайной симмуляции']);
```

Это чёрная магия какая-то. Комбинация двух проигрышных стратегий в итоге даёт победную.
В этом случае, я считаю, это происходит потому что эти две стратегии влияют друг на друга. Например, шанс выиграть три раза подряд при использовании первой стратегии - примерно 11.7%, при использовании второй - 0.5%, но если их удачно скомбинировать, например, на кратное трём богатство выпадет второе казино, а на остальные варианты первое казино, то шансы повышаются до 17.7%. 

Так же, при случайной замене второго казино на первое, матожидание повышается на 18%. Ещё один вариант - возможно две стратегии резонируют друг с другом, снижая разброс значений. Линии на третьем графике значительно кучнее, чем на первых двух.

# Задача 3

```python
n_sim = 100000 #Я решил увеличить выборку, чтобы графики были поровнее
np.random.seed(swag)
random.seed(swag)
n_list = list(range(80, 161, 5))
passed_dudes = []
sad_dudes = []
passed_corr = []
sad_corr = []

for n in n_list:
    passed_dudes_mat_stats = []
    passed_dudes_rus_stats = []
    passed_dudes_mat_rez = []
    passed_dudes_rus_rez = []
    sad_dudes_mat_stats = []
    sad_dudes_rus_stats = []
    sad_dudes_mat_rez = []
    sad_dudes_rus_rez = []
    rus_rez = np.random.normal(60, 10, n_sim)
    mat_rez = np.random.normal(60, 10, n_sim)
    mat_rus_sum = [rus_rez[i]+mat_rez[i] for i in range(n_sim)]
    for i in range(n_sim):
        if mat_rus_sum[i]>n:
            passed_dudes_mat_stats.append(mat_rez[i])
            passed_dudes_rus_stats.append(rus_rez[i])
        else:
            sad_dudes_mat_stats.append(mat_rez[i])
            sad_dudes_rus_stats.append(rus_rez[i])
    passed_dudes.append(len(passed_dudes_mat_stats))
    passed_corr.append(np.corrcoef([passed_dudes_mat_stats, passed_dudes_rus_stats])[0][1])
    sad_corr.append(np.corrcoef([sad_dudes_mat_stats, sad_dudes_rus_stats])[0][1])
        
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 15))
fig.suptitle('', fontsize=22)

ax1.plot(n_list, passed_dudes)
ax1.set_ylabel('Количество прошедших учеников', fontsize=16)
ax1.set_title('Взаимосвязь количества прошедших учеников и проходного балла')

ax2.plot(n_list, passed_corr)
ax2.set_ylabel('Корреляция', fontsize=16)
ax2.set_title('Взаимосвязь проходного балла и корреляции результатов по русскому и математике среди прошедших учеников')

ax3.plot(n_list, sad_corr)
ax3.set_ylabel('Корреляция', fontsize=16)
ax3.set_title('Взаимосвязь проходного балла и корреляции результатов по русскому и математике среди непрошедших учеников')
ax3.set_xlabel('Проходной балл', fontsize=16);
```

Как видно, возникает парадоксальная ситуация: чем выше проходной балл, тем больше отрицательная корреляция между баллами по русскому и математике у прошедших учеников. Это можно объяснить тем, что люди с низкими баллами по обоим предметам точно не смогут пройти. Существует четыре равновероятных варианта: высокий балл по русскому и математике, высокий балл только по русскому, высокий балл только по математике и низкий балл по обоим предметам. В обычной ситуации, если мы берём случайных студентов, то корреляция нулевая, так как все четыре состояния балансируют друг друга. Но если мы берём только прошедших экзамен, то в нашей выборке пропадают ребята получившие плохие баллы по двум предметам. Возникает дисбаланс. Аналогичные результаты мы получаем, если делаем выборку из непрошедших ребят. В таком случае в выборке отсутствуют ребята сдавшие оба предмета хорошо.

P.S. Забавная ситуация возникает, если продолжить понижать проходной балл до 60. Тогда в какой-то момент корреляция результатов у непрошедших ребят дойдёт до -1, а затем резко взлетит до 1. Видимо набрать меньше 60 баллов так тяжело, что парадокс уже не работает
